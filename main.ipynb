# Análisis del Covid-19 en Costa Rica

```python
import pandas as pd
from sklearn.linear_model import LogisticRegressionCV, LinearRegression
```

```python
data = pd.read_csv("data_csv/covid_19_costa_rica.csv")
data.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data["days"], data["total_cases"], 'o')
plt.title("Casos de COVID-19 en Costa Rica")
plt.xlabel("Días transcurridos desde el primer caso confirmado")
plt.ylabel("Número de casos diarios")
```

```python
from sklearn.model_selection import train_test_split
```

```python
#X_train, x_test, Y_train, y_test = train_test_split(data["days"], data["total_cases"])
X_pred = data[["days", "deaths", "recovered"]]
Y = data["total_cases"]
```

```python
lr = LinearRegression()
lr.fit(X_pred, Y)
```

```python
print("lm.intercept: {}\nlm.coef: {}\nlm.score: {}".format(lr.intercept_, lr.coef_, lr.score(X_pred, Y)))
```
